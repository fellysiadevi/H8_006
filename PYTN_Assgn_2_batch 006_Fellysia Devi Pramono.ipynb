# Confidental Interval

```python
datacross = pd.crosstab(new_data.Borough, new_data.Building_Class_at_Present)
datacross
```

```python
Building_Class = datacross.A1[0]
Building_Classes = datacross.A1[1]
total = datacross.A1.sum()
total
```

```python
TotalASatu = Building_Classes/total
TotalASatu
```

```python
TotalZSembilan = datacross.Z9[1] / datacross.Z9.sum() 
TotalZSembilan
```

```python
x = datacross.A1.sum()
x
```

```python
StandarErrorASatu = np.sqrt(TotalASatu * (1 - TotalASatu) / x)
StandarErrorASatu
```

```python
z_score = 1.96
lcb = TotalASatu - z_score*StandarErrorASatu #lower limit of the CI
ucb = TotalASatu + z_score*StandarErrorASatu #upper limit of the CI
lcb, ucb
```

Diperoleh tipe kelas atas (UCB) = 0.13849 dan untuk tipe kelas bawah diperoleh 0.12242 untuk confidental interval dari jenis gedung A1

# Hypothesis Testing

#### Dengan memperhatikan data yang diberikan, apakah jenis gedung A1 dan A3 menentukan sale pricenya juga meningkat?

```python
import statsmodels.api as sm
```

```python
buildingclassA3 = new_data4[new_data4["Building_Class_at_Present"] == 'A3']
buildingclassA1 = new_data4[new_data4["Building_Class_at_Present"] == 'A1']
```

```python
n1 = len(buildingclassA3)
mu1 = buildingclassA3["Sale_Price"].mean()
sd1 = buildingclassA3["Sale_Price"].std()

(n1, mu1, sd1)
```

```python
n2 = len(buildingclassA1)
mu2 = buildingclassA1["Sale_Price"].mean()
sd2 = buildingclassA1["Sale_Price"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(buildingclassA3["Sale_Price"].dropna(), buildingclassA1["Sale_Price"].dropna(),alternative='two-sided')
```

Karena total building class A1 (1.4275e-19) sangat kecil, kita dapat reject the Null hypothesis bahwa rata-rata Price dari building class A3 sama dengan A1. Catatan, kita menggunakan alternative="two-sided" dalam z-test karena di sini kita memeriksa ketidaksamaan.
